# **9.5 Ordinary Differential Equation Solving**

ODE를 풀어야함

1. Euler's Method : 현재 위치의 접선 방향으로 그대로 직진해서 다음지점 예측
 $y_{n+1} = y_n + h f(t_n, y_n)$
 2. Mid-point Method : 현재 위치에서 h/2만큼 간 뒤, 그 중간 지점의 기울기를 구해 처음부터 그 기울기로 다음지점으로 감
 $y_{n+1} = y_n + h f(t_n + \frac{1}{2}h, y_n + \frac{1}{2}h f(t_n, y_n))$
 3. RK4 Method : 4개의 기울기 단계를 가중 평균하여 다음 지점 예측

Second Order differential Equation

$\frac{d^2y}{dt^2}$가 있는 2계방정식은 풀기 까다로움. 두개의 1계 미분 방정식으로 쪼개어 해결

Lotka-Volterra (포식자-피식자 모델)
생태계의 역동적인 균형을 설명하는 비선형 미분 방정식 모델

피식자많아짐 -> 포식자많아짐 -> 피식자줄어듬 -> 포식자줄어듬

SIR모델: 감염병 확산 모델, 인구 집단을 세가지로 나눔

S (Susceptible): 감염될 수 있는 건강한 사람.

I (Infectious): 병을 옮길 수 있는 감염자.

R (Removed): 회복되어 면역이 생겼거나 사망하여 명단에서 제외된 사람.

물리적 의미: 감염률($\beta$)과 회복률($\gamma$)에 의해 그래프가 결정되며, $R_0 = \beta/\gamma$ 값이 전염병의 위력을 나타냄
