# **9.1 Physics Informed Neural Networks using JaxModel & PINN_Model**

PINNs(Physics-Informed Neural Networks)

AI가 물리법칙(미분 방정식)을 배우도록 만드는 방법

일반적인 딥러닝은 정답(y)가 많아야함, But 소재 시뮬레이션 실험값이 별로 없음

idea : 이미 뉴턴 or 나비에-스트로크 방정식 같은 물리 법칙 많이 알고있음 -> 예측값이 이 미분 방정식을 만족하지 못하면 Loss를 주자

Example : Burgers' Equation

경계/초기조건(데이터학습)

물리 법칙 강제(Regularizer)

Total Loss = Initial Loss + Physics Loss(f)

labeled_x,labeled_y 초기조건(t=0)과 경계조건 (x= +-1)을 담고있는 데이터

unlabeled_X : 정답은 모르지만 물리법칙을 만족해야하는 영역 내부의 좌표들

Loss Function

loss_u : 정답데이터와의 오차(데이터 피팅)

loss_f : 내부 좌표들에서 물리 방정식을 얼마나 위반했는지에 대한 오차

각 좌표에서의 오차값들이 하나의 Scalar값으로 합쳐져 최적화에 사용


활성화 함수 : jnp.tanh ( 미분방정식을 풀어야하므로 smooth한 미분이 가능한 tanh 사용, ReLU는 2차 미분이 0이되어 PINN에 부적합)
